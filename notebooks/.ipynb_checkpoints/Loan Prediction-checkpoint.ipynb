There is a meaningful order (e.g., having 2 dependents is more than having 1, and having 3+ is even more), then you should treat this column as ordinal. In this case, using label encoding might be more appropriate.

Though if I use automatic label encoding, 0 dependents will be 0, 1 and 2 will remain the same.3+ dependents will be 3 and missing will be 4 .

This will affect how our model works because many machine learning algorithms assume that higher numerical values indicate a higher importance or value.Treating "unknown" as a higher number could lead to misleading interpretations.

Hence, I will use custome label encoding (Assigning the values myself)

```python
# Custom mapping for label encoding

dependents_mapping = {'0': 0,'1': 1,'2': 2,'3+': 3,'Missing': -1}
X_train_encoded['Dependents'] = X_train_encoded['Dependents'].map(dependents_mapping)
X_test_encoded['Dependents'] = X_test_encoded['Dependents'].map(dependents_mapping)
```

# For evaluation I'm going to consider precision and recall.

## Precision 

Used when false positives are costly.When it's predicted that one will be given a loan and they should not.
Reasons;

1. Minimize Financial Risks - High Precision ensures that when the model predicts a loan approval, it is usually correct.This minimizes the risk of approving loans to individuals who are not likely to default, thus reducing financial loss

2. Building Trust - Ensuring that approved loans are indeed good candidates can build trust with investors and stakeholders,showing that the lending process is reliable and that the financial institution is making sound decisions.

3. Resource Allocation - High precision can help in better resource allocation, focusing efforts on applicants who are likely to repay the loan thus optimizing operational efficiency

## Recall

Used when false negatives are costly(Predicts that one does not deserve a loan when they actually do

So as to cater for both precision and recall, I'll use the harmonic mean of both(F1-Score) because we need to be mindful of both false positives and false negatives.

### BASELINE MODEL (LOGISTIC REGRESSION)

```python
lr = LogisticRegression()

# Training the model using train data
lr.fit(X_train_encoded,y_train)

# Predicting
y_pred = lr.predict(X_test_encoded)

# Evaluating
accuracy = accuracy_score(y_test,y_pred)
precision = precision_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)

# Getting the predicted probabilities
y_pred_proba = lr.predict_proba(X_test_encoded)[:,1]

print(f"Precision: {round((precision * 100),3)}")
print(f"Recall: {round((recall * 100),3)}")
print(f"Accuracy: {round((accuracy * 100),3)}")

print(f"Loss: {round(log_loss(y_test,y_pred_proba)* 100,3)}")
```

From the logistic regression model, 83% of predictions are correct(Accuracy)
Precision - 80% of positive predictions(loan will be approved) that are actually positive.This is important in scenarios where false positives (predicting a loan will be approved when it won't be) are costly.
Recall - There are 98% of actual positives that are correctly predicted.This is important in scenarios where missing an actual loan approval (false negative) is costly.

### NEURAL NETWORKS

```python
import random
import numpy as np

# Setting seed for reproducibility for different libraries to allow reproducibility to be effective

# Creating a function for the sequential model

def sequential_model(neurons=10):
    np.random.seed(42)
    tf.random.set_seed(42)
    random.seed(42)

    model = Sequential()

    # Adding a dense layer
    model.add(Dense(neurons,activation='relu'))
    model.add(Dense(neurons,activation='relu'))
    model.add(Dense(1,activation='sigmoid'))

    # Compiling the model
    model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy',Precision(),Recall()])

    # Train the model
    history = model.fit(X_train_encoded,y_train,epochs=50,batch_size=32)

    # Evaluate the model
    evaluation = model.evaluate(X_train_encoded,y_train)

    loss = round(evaluation[0]*100,2)
    accuracy = round(evaluation[1]*100,2)
    precision = round(evaluation[2]*100,2)
    recall = round(evaluation[3]*100,2)

    return f"Loss: {loss}, Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}"
```

```python
# Calling the function

sequential_model()
```

### According to the above performance, these are the comments

Loss: 44.29: This is the binary cross-entropy loss, which measures how well the model's predicted probabilities match the actual class labels. A lower loss indicates better performance.

Accuracy: 81.6: This indicates that 81.6% of the predictions made by the model are correct.

Precision: 80.81: This indicates that 80.81% of the instances predicted as positive (loan approval) are actually positive.

Recall: 96.14%: This indicates that 96.14% of the actual positive instances (loan approvals) are correctly identified by the model.

```python
param_grid = {
    'n_estimators': [50,100,200,300],
    'max_depth': [None,10,20,30,40,50],  # Maximum depth of the tree
    'min_samples_split': [2,5,10],  # Minimum number of samples required to split at an internal node
    'min_samples_leaf': [1,2,4]  # Min number of samples to be at a leaf node
}
# Instantiating the classifier
rf = RandomForestClassifier(random_state=42)

random_search = RandomizedSearchCV(estimator=rf,param_distributions=param_grid,n_iter=50,cv=3,random_state=42)

# Fitting
random_search.fit(X_train_encoded,y_train)

# Predict using the best model
random_search_pred = random_search.predict(X_test_encoded)

best_params = random_search.best_params_





# Fitting the best parameters

rf_best = RandomForestClassifier(**best_params,random_state=42)

# Training the model now using best parameters
rf_best.fit(X_train_encoded,y_train)

# Dumping the variables i'll need in a joblib file
joblib.dump(X_train_encoded,'X_train_encoded.pkl')
joblib.dump(X_test_encoded,'X_test_encoded.pkl')
joblib.dump(y_train,'y_train.pkl')
joblib.dump(y_test,'y_test.pkl')

# Serializing the model
joblib.dump(rf_best, 'rf_best.pkl')




# Predict using the refitted model
# rf_best_pred = rf_best.predict(X_test_encoded)

# Evaluation of our model
# print(f"Precision: {round(precision_score(y_test,rf_best_pred)*100,2)}")
# print(f"Recall: {round(recall_score(y_test,rf_best_pred)*100,2)}")
# print(f"Accuracy: {round(accuracy_score(y_test,rf_best_pred)*100,2)}")
```
